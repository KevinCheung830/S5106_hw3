```python
# Compute average delay by destination
avg_delay_by_dest = flights.groupby('dest').agg({
    'arr_delay': 'mean',
    'dep_delay': 'mean',
    'flight': 'count'
}).reset_index()
avg_delay_by_dest.columns = ['dest', 'avg_arr_delay', 'avg_dep_delay', 'flight_count']

# Join with airports data
airports_data = airports[['faa', 'lat', 'lon', 'name', 'tz', 'dst']]
avg_delay_with_airports = avg_delay_by_dest.merge(
    airports_data, 
    left_on='dest', 
    right_on='faa', 
    how='inner'
)

# Remove rows with missing coordinates
avg_delay_with_airports = avg_delay_with_airports.dropna(subset=['lat', 'lon', 'avg_arr_delay'])

print("1. AVERAGE DELAY BY DESTINATION")
print("Top 10 destinations with highest average arrival delays:")
print(avg_delay_with_airports.nlargest(10, 'avg_arr_delay')[['name', 'avg_arr_delay', 'flight_count']].to_string(index=False))

# Try Plotly first, if not available use matplotlib
try:
    import plotly.express as px
    fig1 = px.scatter_geo(
        avg_delay_with_airports,
        lat='lat',
        lon='lon',
        size='flight_count',
        color='avg_arr_delay',
        hover_name='name',
        hover_data={
            'avg_arr_delay': ':.2f',
            'avg_dep_delay': ':.2f',
            'flight_count': True,
            'lat': False,
            'lon': False
        },
        title='Spatial Distribution of Average Arrival Delays by Destination',
        color_continuous_scale='RdBu_r',
        color_continuous_midpoint=0
    )
    fig1.show()
    
except ImportError:
    print("Plotly not available, using matplotlib instead...")
    # Fallback to matplotlib
    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(
        avg_delay_with_airports['lon'],
        avg_delay_with_airports['lat'],
        s=avg_delay_with_airports['flight_count']/10,
        c=avg_delay_with_airports['avg_arr_delay'],
        cmap='RdBu_r',
        alpha=0.7,
        vmin=-20, vmax=20
    )
    plt.colorbar(scatter, label='Average Arrival Delay (minutes)')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Spatial Distribution of Average Arrival Delays by Destination')
    
    # Add some airport labels
    for i, row in avg_delay_with_airports.nlargest(5, 'avg_arr_delay').iterrows():
        plt.annotate(row['dest'], (row['lon'], row['lat']), xytext=(5, 5), 
                    textcoords='offset points', fontsize=8)
    
    plt.tight_layout()
    plt.show()
```

```python
print("\n2. RELATIONSHIP BETWEEN PLANE AGE AND DELAYS")

# Prepare planes data
planes_data = planes[['tailnum', 'year']].copy()
planes_data['plane_age'] = 2013 - planes_data['year']

# Merge with flights data
flights_with_age = flights.merge(planes_data, on='tailnum', how='inner')

# Remove unrealistic ages
flights_with_age = flights_with_age[
    (flights_with_age['plane_age'] >= 0) & 
    (flights_with_age['plane_age'] <= 60) &
    (flights_with_age['arr_delay'].notna())
]

# Calculate average delays by plane age
age_delay_relationship = flights_with_age.groupby('plane_age').agg({
    'arr_delay': 'mean',
    'dep_delay': 'mean',
    'flight': 'count'
}).reset_index()

# Remove age groups with too few flights
age_delay_relationship = age_delay_relationship[age_delay_relationship['flight'] > 10]

# Create visualization
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Arrival delay vs plane age
scatter1 = ax1.scatter(
    age_delay_relationship['plane_age'],
    age_delay_relationship['arr_delay'],
    s=age_delay_relationship['flight']/50,
    c=age_delay_relationship['arr_delay'],
    cmap='viridis',
    alpha=0.7
)
ax1.set_xlabel('Plane Age (years)')
ax1.set_ylabel('Average Arrival Delay (minutes)')
ax1.set_title('Arrival Delay vs Plane Age')
plt.colorbar(scatter1, ax=ax1, label='Arrival Delay (minutes)')

# Departure delay vs plane age
scatter2 = ax2.scatter(
    age_delay_relationship['plane_age'],
    age_delay_relationship['dep_delay'],
    s=age_delay_relationship['flight']/50,
    c=age_delay_relationship['dep_delay'],
    cmap='viridis',
    alpha=0.7
)
ax2.set_xlabel('Plane Age (years)')
ax2.set_ylabel('Average Departure Delay (minutes)')
ax2.set_title('Departure Delay vs Plane Age')
plt.colorbar(scatter2, ax=ax2, label='Departure Delay (minutes)')

plt.tight_layout()
plt.show()

# Statistical analysis
correlation_arr = age_delay_relationship['plane_age'].corr(age_delay_relationship['arr_delay'])
correlation_dep = age_delay_relationship['plane_age'].corr(age_delay_relationship['dep_delay'])

print(f"Correlation between plane age and arrival delay: {correlation_arr:.3f}")
print(f"Correlation between plane age and departure delay: {correlation_dep:.3f}")

# Additional analysis by age groups
age_delay_relationship['age_group'] = pd.cut(age_delay_relationship['plane_age'], 
                                           bins=[0, 5, 10, 15, 20, 60])
age_group_stats = age_delay_relationship.groupby('age_group').agg({
    'arr_delay': 'mean',
    'dep_delay': 'mean',
    'flight': 'sum'
}).reset_index()

print("\nAverage delays by plane age groups:")
print(age_group_stats.to_string(index=False))
```

```python
import seaborn as sns
print("\n3. WEATHER CONDITIONS AND DELAYS")

# Merge flights with weather data
flights_weather = flights.merge(
    weather,
    on=['year', 'month', 'day', 'hour', 'origin'],
    how='inner'
)

# Select relevant weather columns and delays
weather_columns = ['temp', 'dewp', 'humid', 'wind_dir', 'wind_speed', 'wind_gust', 'precip', 'pressure', 'visib']
weather_delay_analysis = flights_weather[weather_columns + ['arr_delay', 'dep_delay']].dropna()

# Calculate correlations
correlation_matrix = weather_delay_analysis.corr()

# Create correlation heatmap
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(correlation_matrix[['arr_delay', 'dep_delay']].drop(['arr_delay', 'dep_delay']), dtype=bool))
sns.heatmap(
    correlation_matrix[['arr_delay', 'dep_delay']].drop(['arr_delay', 'dep_delay']),
    annot=True,
    cmap='RdBu_r',
    center=0,
    mask=mask,
    fmt='.3f'
)
plt.title('Correlation Between Weather Conditions and Flight Delays')
plt.tight_layout()
plt.show()

# Analyze specific weather conditions
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Wind speed vs delays
axes[0,0].hexbin(weather_delay_analysis['wind_speed'], weather_delay_analysis['arr_delay'], 
                gridsize=30, cmap='Blues', alpha=0.7)
axes[0,0].set_xlabel('Wind Speed (mph)')
axes[0,0].set_ylabel('Arrival Delay (minutes)')
axes[0,0].set_title('Wind Speed vs Arrival Delay')

# Precipitation vs delays
precip_data = weather_delay_analysis[weather_delay_analysis['precip'] > 0]
axes[0,1].hexbin(precip_data['precip'], precip_data['arr_delay'], 
                gridsize=20, cmap='Reds', alpha=0.7)
axes[0,1].set_xlabel('Precipitation (inches)')
axes[0,1].set_ylabel('Arrival Delay (minutes)')
axes[0,1].set_title('Precipitation vs Arrival Delay (>0 precip only)')

# Visibility vs delays
axes[1,0].hexbin(weather_delay_analysis['visib'], weather_delay_analysis['arr_delay'], 
                gridsize=30, cmap='Greens', alpha=0.7)
axes[1,0].set_xlabel('Visibility (miles)')
axes[1,0].set_ylabel('Arrival Delay (minutes)')
axes[1,0].set_title('Visibility vs Arrival Delay')

# Humidity vs delays
axes[1,1].hexbin(weather_delay_analysis['humid'], weather_delay_analysis['arr_delay'], 
                gridsize=30, cmap='Purples', alpha=0.7)
axes[1,1].set_xlabel('Humidity (%)')
axes[1,1].set_ylabel('Arrival Delay (minutes)')
axes[1,1].set_title('Humidity vs Arrival Delay')

plt.tight_layout()
plt.show()

# Statistical summary
print("Average delays by weather conditions:")
print(f"High wind (>20 mph): {weather_delay_analysis[weather_delay_analysis['wind_speed'] > 20]['arr_delay'].mean():.2f} min")
print(f"Low wind (≤20 mph): {weather_delay_analysis[weather_delay_analysis['wind_speed'] <= 20]['arr_delay'].mean():.2f} min")
print(f"With precipitation: {weather_delay_analysis[weather_delay_analysis['precip'] > 0]['arr_delay'].mean():.2f} min")
print(f"No precipitation: {weather_delay_analysis[weather_delay_analysis['precip'] == 0]['arr_delay'].mean():.2f} min")
print(f"Low visibility (<5 miles): {weather_delay_analysis[weather_delay_analysis['visib'] < 5]['arr_delay'].mean():.2f} min")
print(f"Good visibility (≥5 miles): {weather_delay_analysis[weather_delay_analysis['visib'] >= 5]['arr_delay'].mean():.2f} min")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the medals data
medals = pd.read_csv('./data/medals.csv')

#print("First few rows of medals data:")
#print(medals.head())
#print("\nMedal types in dataset:")
#print(medals['medal_type'].value_counts())
#print("\nCountries in dataset:")
#print(medals['country'].value_counts().head(10))

# Count medals by country and type
medal_counts = medals.groupby(['country', 'medal_type']).size().unstack(fill_value=0)

#print(f"\nMedal counts shape: {medal_counts.shape}")
#print("Medal counts columns:", medal_counts.columns.tolist())

# Ensure we have all medal types
for medal_type in ['Gold Medal', 'Silver Medal', 'Bronze Medal']:
    if medal_type not in medal_counts.columns:
        medal_counts[medal_type] = 0

# Reorder columns to be Gold, Silver, Bronze
medal_counts = medal_counts[['Gold Medal', 'Silver Medal', 'Bronze Medal']]

# Calculate total medals for ranking
medal_counts['total'] = medal_counts.sum(axis=1)

#print(f"\nCountries with medals: {len(medal_counts[medal_counts['total'] > 0])}")

# Sort by total medals (rank) and get top 30 countries
top_30_medals = medal_counts.nlargest(30, 'total')

# print(f"\nTop 30 countries by total medals:")
# print(top_30_medals[['Gold Medal', 'Silver Medal', 'Bronze Medal', 'total']].sort_values('total', ascending=False))

# Remove total column for plotting
plot_data = top_30_medals.drop('total', axis=1)

# Define colors for medal types
medal_colors = ['#FFD700', '#C0C0C0', '#CD7F32']  # Gold, Silver, Bronze

# Create the stacked bar chart
plt.figure(figsize=(16, 10))
ax = plot_data.plot(kind='bar', 
                   stacked=True, 
                   color=medal_colors,
                   figsize=(16, 10))

plt.title('Medal Counts by Country (Top 30 Nations)', 
          fontsize=16, 
          fontweight='bold', 
          pad=20)
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Number of Medals', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Customize legend with correct medal names
plt.legend(['Gold Medal', 'Silver Medal', 'Bronze Medal'], 
           title='Medal Type', 
           title_fontsize=12,
           fontsize=11,
           frameon=True,
           fancybox=True,
           shadow=True,
           loc='upper right')

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, 
                 label_type='center', 
                 fontsize=9, 
                 color='black',
                 fontweight='bold',
                 padding=2)

# Add grid for better readability
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the chart
plt.show()

# Print the final ranking table
print("\n" + "="*70)
print("FINAL RANKING - Top 30 Countries by Medal Count")
print("="*70)
final_ranking = top_30_medals[['Gold Medal', 'Silver Medal', 'Bronze Medal', 'total']].sort_values('total', ascending=False)
print(final_ranking)

print(f"\nSummary Statistics:")
print(f"Total medals awarded: {medal_counts['total'].sum()}")
print(f"Total countries with medals: {len(medal_counts[medal_counts['total'] > 0])}")
print(f"Gold medals: {medal_counts['Gold Medal'].sum()}")
print(f"Silver medals: {medal_counts['Silver Medal'].sum()}")
print(f"Bronze medals: {medal_counts['Bronze Medal'].sum()}")
medal_colors = ['#FFD700', '#C0C0C0', '#CD7F32']  # Gold, Silver, Bronze
```

# Question 4 Pi estimation


```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
num_needles = 100000
num_experiments = 100
d = 1  # line spacing
l = 1  # needle length

# Store pi estimates
pi_estimates = []

for exp in range(num_experiments):
    # Part A: Generate random x positions and theta angles
    x = np.random.uniform(0, 1, num_needles)  # distance from left bar [0, 1]
    theta = np.random.uniform(0, 2 * np.pi, num_needles)  # orientation [0, 2π]
    
    # Part B: Calculate needle end positions
    x_end1 = x + (l/2) * np.cos(theta)  # right end
    x_end2 = x - (l/2) * np.cos(theta)  # left end
    
    # Find min and max x positions for each needle
    x_end_min = np.minimum(x_end1, x_end2)
    x_end_max = np.maximum(x_end1, x_end2)
    
    # Part C: Count needles that cross the vertical lines
    # A needle crosses if one end is below 0 and the other above 0
    # (Since lines are at integer positions: ..., -1, 0, 1, 2, ...)
    crosses_line = np.logical_and(x_end_min < 0, x_end_max > 0)
    
    # Estimate pi from crossing probability
    crossing_prob = np.sum(crosses_line) / num_needles
    pi_estimate = 1 / crossing_prob  # Since P(cross) = 1/π when d=l=1
    pi_estimates.append(pi_estimate)

# Print one example estimate
print(pi_estimates[0])

# Part D: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, 'b-', alpha=0.7, linewidth=0.8)
plt.axhline(y=np.pi, color='r', linestyle='--', label='True π')
plt.xlabel('Experiment Number')
plt.ylabel('Estimated π value')
plt.title('Buffon\'s Needle: Monte Carlo Estimation of π')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Print statistics
print(f"Mean π estimate: {np.mean(pi_estimates):.6f}")
print(f"Std deviation: {np.std(pi_estimates):.6f}")
print(f"True π: {np.pi:.6f}")
```
