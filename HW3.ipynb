```python
# 5. Arrived >2 hours late but didn't leave late
result_a5 = df_flight[(df_flight['arr_delay'] > 120) & (df_flight['dep_delay'] <= 0)]
print(f"5. Arrived >2h late but didn't leave late: {len(result_a5)} flights")
print(result_a5.head())
```

```python
# 6. Made up time in flight
result_a6 = df_flight[(df_flight['dep_delay'] >= 60) & (df_flight['dep_delay'] - df_flight['arr_delay'] > 30)]
print(f"6. Made up >30min in flight: {len(result_a6)} flights")
print(result_a6.head())
```

```python
# 7. Arrived between midnight and 6am
def is_early_arrival(time):
    if pd.isna(time):
        return False
    time_int = int(time)
    return (0 <= time_int <= 600) or (time_int == 2400)

result_a7 = df_flight[df_flight['arr_time'].apply(is_early_arrival)]
print(f"7. Arrived between midnight and 6am: {len(result_a7)} flights")
print(result_a7.head())
```

```python
df = flights

print("=== Summary for all PART A ===")

# 1. Departure delay of two or more hours
result1 = df[df['dep_delay'] >= 120]
print(f"1. Flights with departure delay ≥ 2 hours: {len(result1)} flights")

# 2. Flew to Houston
result2 = df[df['dest'].isin(['IAH', 'HOU'])]
print(f"2. Flights to Houston: {len(result2)} flights")

# 3. United, American, or Delta
result3 = df[df['carrier'].isin(['UA', 'AA', 'DL'])]
print(f"3. Flights operated by United, American, or Delta: {len(result3)} flights")

# 4. Winter flights
result4 = df[df['month'].isin([12, 1, 2])]
print(f"4. Winter flights: {len(result4)} flights")

# 5. Arrived >2 hours late but didn't leave late
result5 = df[(df['arr_delay'] > 120) & (df['dep_delay'] <= 0)]
print(f"5. Arrived >2h late but didn't leave late: {len(result5)} flights")

# 6. Made up time in flight
result6 = df[(df['dep_delay'] >= 60) & (df['dep_delay'] - df['arr_delay'] > 30)]
print(f"6. Made up >30min in flight: {len(result6)} flights")

# 7. Arrived between midnight and 6am
def is_early_arrival(time):
    if pd.isna(time):
        return False
    time_int = int(time)
    return (0 <= time_int <= 600) or (time_int == 2400)

result7 = df[df['arr_time'].apply(is_early_arrival)]
print(f"7. Arrived between midnight and 6am: {len(result7)} flights")

print("\n=== PART B ===")

# Most arrival delayed flights
most_delayed = df[['flight', 'carrier', 'origin', 'dest', 'arr_delay', 'dep_delay']].sort_values('arr_delay', ascending=False).head(10)
print("Most arrival delayed flights:")
print(most_delayed.to_string(index=False))

# Flights that landed earliest
earliest_landed = df[['flight', 'carrier', 'origin', 'dest', 'arr_delay', 'dep_delay']].sort_values('arr_delay', ascending=True).head(10)
print("\nFlights that landed earliest (smallest delay):")
print(earliest_landed.to_string(index=False))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
num_needles = 100000
num_experiments = 100
d = 1  # line spacing
l = 1  # needle length

# Store pi estimates
pi_estimates = []

for exp in range(num_experiments):
    # Part A: Generate random x positions and theta angles
    x = np.random.uniform(0, 1, num_needles)  # distance from left bar [0, 1]
    theta = np.random.uniform(0, 2 * np.pi, num_needles)  # orientation [0, 2π]
    
    # Part B: Calculate needle end positions
    x_end1 = x + (l/2) * np.cos(theta)  # right end
    x_end2 = x - (l/2) * np.cos(theta)  # left end
    
    # Find min and max x positions for each needle
    x_end_min = np.minimum(x_end1, x_end2)
    x_end_max = np.maximum(x_end1, x_end2)
    
    # Part C: Count needles that cross the vertical lines
    # A needle crosses if one end is below 0 and the other above 0
    # (Since lines are at integer positions: ..., -1, 0, 1, 2, ...)
    crosses_line = np.logical_and(x_end_min < 0, x_end_max > 0)
    
    # Estimate pi from crossing probability
    crossing_prob = np.sum(crosses_line) / num_needles
    pi_estimate = 1 / crossing_prob  # Since P(cross) = 1/π when d=l=1
    pi_estimates.append(pi_estimate)

# Print one example estimate
print(pi_estimates[0])

# Part D: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(pi_estimates, 'b-', alpha=0.7, linewidth=0.8)
plt.axhline(y=np.pi, color='r', linestyle='--', label='True π')
plt.xlabel('Experiment Number')
plt.ylabel('Estimated π value')
plt.title('Buffon\'s Needle: Monte Carlo Estimation of π')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Print statistics
print(f"Mean π estimate: {np.mean(pi_estimates):.6f}")
print(f"Std deviation: {np.std(pi_estimates):.6f}")
print(f"True π: {np.pi:.6f}")
```
